```python
#check for null
shl_df.isnull().sum()
```

```python
#check values for BMI Category 
shl_df['BMI Category'].value_counts()
```

```python
#compare relationship between BMI and sleep disorder
shl_df.groupby('BMI Category')['Sleep Disorder'].value_counts()
```

```python
#check for number of people having sleep disorder
shl_df['Sleep Disorder'].value_counts()
```

```python
#check sleep disorder genderwise
shl_df.groupby('Sleep Disorder')['Gender'].value_counts()
```

```python
#The effect of job on sleep
shl_df.groupby('Sleep Disorder') ['Occupation'].value_counts()
```

```python
#physical activity
shl_df.groupby('Age')['Sleep Disorder'].value_counts()
```

```python
#check for quality of sleep between male and female
shl_df.groupby('Gender')['Quality of Sleep'].value_counts()
```

```python
#check for sleep duration
shl_df.groupby('Gender')['Sleep Duration'].value_counts()
```

```python
#check for stress level
shl_df.groupby('Stress Level')['Sleep Disorder'].value_counts()
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
shl_df.pivot_table(index='Gender',columns='Sleep Disorder',values='Age',aggfunc='median').plot(kind='bar',color={'#FFF3FD','#4A235A','pink'},
                                                                                                   title='Most affected ages in each type of Sleep Disorder',
                                                                                                    label='Age',alpha=.7)


plt.show()
```

```python
#visualizing the correlation features
plt.figure(figsize=(10, 8))
sns.heatmap(shl_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Visualizing relationships
sns.scatterplot(data=shl_df, x="Sleep Duration", y="Quality of Sleep", hue="Gender")
plt.show()

# Comparing data across groups
grouped_data = shl_df.groupby("Stress Level")["Sleep Duration"].mean()
```
